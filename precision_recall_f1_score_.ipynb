```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
```

```python
heart=pd.read_csv("/content/heart.csv")
heart.head()
```

```python
X=heart.drop(columns='target',axis=1)
Y=heart['target']
#
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
heart_2=heart.drop(columns='target',axis=1)
heart_2=heart_2.drop(columns='fbs',axis=1)
heart_2=heart_2.drop(columns='restecg',axis=1)
heart_2=heart_2.drop(columns='exang',axis=1)
heart_2=heart_2.drop(columns='slope',axis=1)
heart_2=heart_2.drop(columns='ca',axis=1)
heart_2=heart_2.drop(columns='thal',axis=1)
heart_2.head()
sclar=StandardScaler()
heart_2=sclar.fit_transform(heart_2)
heart_2=pd.DataFrame(heart_2)
heart_2.head()
```

```python
heart_features = heart.drop(columns='target').replace(0, np.nan)
heart_features = heart_features.fillna(heart_features.mean())

# Now replace the scaled features back, but keep the original target column
heart['age'] = heart_2[0]
heart['trestbps'] = heart_2[1]
heart['chol'] = heart_2[2]
heart['thalach'] = heart_2[3]
heart['oldpeak'] = heart_2[4]

# Make sure the target column is not affected by the imputation
heart['target'] = heart['target'].replace(np.nan, 0) # Revert the 0s back if they were changed

print(heart.head())
```

```python
X=heart.drop(columns="target",axis=1)
Y=heart['target']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
model.fit(X_train,Y_train)
```

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
print("Accuracy on training data:",training_data_accuracy)
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
print("Accuracy on test data:",test_data_accuracy)
```

**Precision** - Ratio of number of true positive to the total number of predicted positive .

```python
from sklearn.metrics import precision_score
```

```python
#precision for training data prediction
precision_train=precision_score(X_train_prediction,Y_train)
```

```python
print(precision_train)
```

```python
#precision for test data prediction
precision_test=precision_score(X_test_prediction,Y_test)
print(precision_test)
```

**Recall** - ratio of true positive to the actual number of positives

```python
from sklearn.metrics import recall_score
#recall for training data prediction
recall_train=recall_score(X_train_prediction,Y_train)
print(recall_train)
#recall for test data prediction
recall_test=recall_score(X_test_prediction,Y_test)
print(recall_test)
```

**F1 Score**-It is the harmonic mean of precision and recall

```python
from sklearn.metrics import f1_score
#f1 for training data prediction
f1_train=f1_score(X_train_prediction,Y_train)
print(f1_train)
#f1 for test data prediction
f1_test=f1_score(X_test_prediction,Y_test)
print(f1_test)
```

```python
def precision_recall_f1(y_true, y_pred):
    precision = precision_score(y_true, y_pred)
    recall = recall_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred)

    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")
```

```python
#classification for training data
precision_recall_f1(X_train_prediction,Y_train)
```

```python
#classification for testing data
precision_recall_f1(X_test_prediction,Y_test)
```
