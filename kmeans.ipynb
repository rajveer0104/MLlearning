```python

import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/income.csv')
```

```python
plt.scatter(df['Age'],df['Income($)'])
plt.plot
```

```python
km=KMeans(n_clusters=3)
km
```

```python
y_pred=km.fit_predict(df[['Age','Income($)']])
y_pred
```

```python
df['Cluster']=y_pred
df.head()
```

```python
df1=df[df.Cluster==0]
df2=df[df.Cluster==1]
df3=df[df.Cluster==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
scaler=MinMaxScaler()
df['Income($)']=scaler.fit_transform(df[['Income($)']])
df['Age']=scaler.fit_transform(df[['Age']])
df.head()
```

```python
y_pred_2=km.fit_predict(df[['Age','Income($)']])
y_pred_2
```

```python
df["Cluster"]=y_pred_2
df.head()
```

```python
df1=df[df['Cluster']==0]
df2=df[df['Cluster']==1]
df3=df[df['Cluster']==2]

plt.scatter(df1['Age'],df1['Income($)'],color='green')
plt.scatter(df2['Age'],df2['Income($)'],color='red')
plt.scatter(df3['Age'],df3['Income($)'],color='black')
plt.xlabel('Age')
plt.ylabel('Income($)')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.legend()
```

```python
k_range=range(1,10)
sse=[]
for k in k_range:
  km=KMeans(n_clusters=k)
  km.fit(df[['Age','Income($)']])
  sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel("sse")
plt.plot(k_range,sse)
```

```python
df5=pd.read_csv('/content/Placement_Dataset.csv')
df5.head()
```

```python
plt.scatter(df5['ssc_p'],df5['etest_p'])
plt.show()
```

```python
scaler=MinMaxScaler()
df5['ssc_p']=scaler.fit_transform(df5[['ssc_p']])
df5['etest_p']=scaler.fit_transform(df5[['etest_p']])
```

```python
df5.head()
```

```python
plt.scatter(df5['ssc_p'],df5['etest_p'])
plt.show()
```

```python
km=KMeans(n_clusters=5)
km
```

```python
ym=km.fit_predict(df5[['ssc_p','etest_p']])
```

```python
df5['Clusters']=ym
```

```python
df5.head()
```

```python
df5_1=df5[df5['Clusters']==0]
df5_2=df5[df5['Clusters']==1]
df5_3=df5[df5['Clusters']==2]
df5_4=df5[df5['Clusters']==3]
df5_5=df5[df5['Clusters']==4]


plt.scatter(df5_1['ssc_p'],df5_1['etest_p'],color='green')
plt.scatter(df5_2['ssc_p'],df5_2['etest_p'],color='blue')
plt.scatter(df5_3['ssc_p'],df5_3['etest_p'],color='red')
plt.scatter(df5_4['ssc_p'],df5_4['etest_p'],color='black')
plt.scatter(df5_5['ssc_p'],df5_5['etest_p'],color='yellow')
plt.xlabel('ssc_p')
plt.ylabel('etest_p')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='x')

plt.show()
```

```python
k=range(1,11)
sse=[]
for m in k:
  p=KMeans(n_clusters=m)
  p.fit_predict(df5[['ssc_p','etest_p']])
  sse.append(p.inertia_)
```

```python
plt.scatter(k,sse)
plt.show()
```
