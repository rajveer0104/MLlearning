```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
```

```python
#loading the breast cancer dataset
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
dataset=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
```

```python
dataset.head()
```

```python
dataset['label']=breast_cancer_dataset.target
```

```python
dataset['label'].value_counts()
```

1- Beningn                
2- Malignant

```python
X=dataset.drop(columns='label',axis=1)
Y=dataset['label']
```

```python
print(X)
print(Y)
```

```python
X=np.asarray(X)
Y=np.asarray(Y)
```

```python
print(X)
print(Y)
```

#**Grid Search CV**   
Grid search is used to determine best parameters for our model


```python
model=SVC()
```

```python
#hyperparameters
parameters ={
    "kernel":['linear','poly','rbf','sigmoid'],

    'C':[1,5,10,20]
}
```

```python
#grid search
classifier=GridSearchCV(model,parameters,cv=5)
```

```python
#fitting data to model
classifier.fit(X,Y)
```

```python
classifier.cv_results_
```

```python
#best parameters
best=classifier.best_params_
print(best)
#best accuracy
best_accuracy=classifier.best_score_
print(best_accuracy)
```

```python
#loading result to pandas dataframe
results=pd.DataFrame(classifier.cv_results_)
```

```python
results.head()
```

```python
grid_search_result = results[['param_C','param_kernel','mean_test_score']]
```

```python
print(grid_search_result)
```

#Randomized search cv

```python
model=SVC()
#hyperparameters
parameters ={
    "kernel":['linear','poly','rbf','sigmoid'],

    'C':[1,5,10,20]
}
```

```python
#randomized search
classifier=RandomizedSearchCV(model,parameters,cv=5)
```

```python
classifier.fit(X,Y)
```

```python
classifier.cv_results_
#best parameters
best=classifier.best_params_
print(best)
#best accuracy
best_accuracy=classifier.best_score_
print(best_accuracy)
```

```python
#loading result to pandas dataframe
results=pd.DataFrame(classifier.cv_results_)
results.head()
```

```python
randomized_search_result = results[['param_C','param_kernel','mean_test_score']]
print(randomized_search_result)
```
